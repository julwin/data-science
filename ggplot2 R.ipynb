# state.x77 dataset

```r
library(ggplot2)
```

```r
df1 = data.frame(state.x77)
```

```r
ggplot(data=df1)+geom_point(mapping=aes(x=Life.Exp, y=Murder))
```

```r
head(df1,10)
```

# Scatter plot with color and size mapping

```r
ggplot(data=df1)+geom_point(mapping=aes(x=Life.Exp, y=Murder, size=Illiteracy, color=HS.Grad))
```

## Finding a linear relationship

```r
cbPalette <- c("#999999", "#E69F00", "#56B4E9", "#009E73", "#F0E442", "#0072B2", "#D55E00", "#CC79A7")
```

```r
model1 <- lm(Murder~Life.Exp, data=df1)
```

```r
fitted1 <- data.frame(lifeexp=df1$Life.Exp, predict(model1, interval='confidence'))
```

```r
head(fitted1,10)
```

## Adding the fitted model to the plot

```r
ggplot(data=df1)+geom_point(mapping=aes(x=Life.Exp, y=Murder, size=Illiteracy, color=HS.Grad))+geom_line(data=fitted1, mapping=aes(x=lifeexp, y=fit, color=50))
```

We can see in the above visualization that there seems to be a linear correlation between Murder/Life experience and Illiteracy / HS.grad

## Comparing "horsepower" with a barplot

```r
ggplot(data=mtcars, mapping=aes(x=rownames(mtcars), y=hp))+geom_bar(stat='identity')+theme(axis.text.x=element_text(angle=90, hjust=1, vjust=0.3))
```

## Coloring the bars by mapping "miles per gallon"

```r
ggplot(data=mtcars, mapping=aes(x=rownames(mtcars), y=hp, color=mpg, fill=mpg))+geom_bar(stat='identity')+theme(axis.text.x=element_text(angle=90, hjust=1, vjust=0.3))
```

## Changing the fill

```r
help(scale_fill_gradient)
```

```r
ggplot(data=mtcars, mapping=aes(x=rownames(mtcars), y=hp, fill=mpg))+geom_bar(stat='identity')+theme(axis.text.x=element_text(angle=90, hjust=1, vjust=0.3))+scale_fill_gradient(low='yellow', high='indianred')
```

```r
ggplot(data=mtcars, mapping=aes(x=rownames(mtcars), y=hp, colour=mpg, fill=mpg))+geom_bar(stat='identity')+theme(axis.text.x=element_text(angle=90, hjust=1, vjust=0.3))+scale_colour_gradient(low='yellow', high='red')+scale_fill_gradient(low='grey', high='black')
```
