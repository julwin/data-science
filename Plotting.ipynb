# Annotations

```python
random.seed(42)
df5 = DataFrame({'one': random.randn(10)**2, 'two': random.randn(10)**2, 'three': random.randn(10)**2})
```

```python
df5
```

```python
fig = plt.figure()
```

```python
ax = fig.add_subplot(1,1,1)
```

```python
ax.fill_between(x=range(len(df5)), y1 = df5.one, facecolor='yellowgreen')
```

```python
ax.fill_between(x=range(len(df5)), y1 = df5.two, facecolor='forestgreen')
```

```python
ax.fill_between(x=range(len(df5)), y1 = df5.three, facecolor='lawngreen')
```

```python
ax.annotate('Test-Text', xy=(3,2.03), xytext=(5,3.5), arrowprops=dict(facecolor='black'))
```

```python
plt.show()
```

```python
display(fig)
```
